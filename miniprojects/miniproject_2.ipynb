```python
# task 5
# Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.

bookings.columns = bookings.columns.str.replace(' ', '_')
bookings.columns = map(str.lower, bookings.columns)
bookings
```

```python
# task 6
# Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

# bookings.loc[bookings.is_canceled == 0].country.value_counts()

bookings_success = bookings.query("is_canceled == 0").country.value_counts().head(7)
bookings_success
```

```python
# task 7
# На сколько ночей в среднем бронируют отели разных типов?

avg_night = bookings.groupby('hotel').stays_total_nights.mean().round(2)
avg_night
```

```python
# task 8
# Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type). Такое может произойти, например, по причине овербукинга. 
#Сколько подобных наблюдений встретилось в датасете?

# b_dif = bookings.query("reserved_room_type != assigned_room_type")
# len(b_dif)
b_dif = bookings.query("reserved_room_type != assigned_room_type").count()
b_dif
```

```python
# task 9
#Проанализируйте даты запланированного прибытия. – На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?

bk_month = bookings.groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
                   .agg({'arrival_full_date': 'count'})  \
                   .sort_values(['arrival_date_year', 'arrival_date_month'], ascending=False)
bk_month
```

```python
# task 9

bookings.groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
        .agg({'lead_time': 'count'}) \
        .sort_values(['arrival_date_year', 'lead_time'], ascending=False)
```

```python
# task 10 
# #– Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов

bk_cansel = bookings.groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
                    .agg({'is_canceled': 'sum'}) \
                    .sort_values('is_canceled', ascending=False).head(10)
bk_cansel
```

```python
# task 11
# Посмотрите на числовые характеристики трёх переменных: adults, children и babies. Какая из них имеет наибольшее среднее значение?

avg_age = bookings[['adults', 'children', 'babies']].mean()
avg_age
```

```python
# task 12
# Создайте колонку total_kids, объединив children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

bookings['total_kids'] = bookings['children'] + bookings['babies']
avg_hotel = bookings.groupby('hotel', as_index=False).agg({'total_kids': 'mean'}).round(2)
#bookings.groupby('hotel').total_kids.mean()
avg_hotel
```

```python
# task 13
# Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. 

def has_kids(x):
    if x >= 1:
        return True
    else:
        return False

bookings['has_kids'] = bookings['total_kids'].apply(has_kids)  
bookings['has_kids']
```
